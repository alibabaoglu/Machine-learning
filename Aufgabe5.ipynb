```python
import numpy as np
import matplotlib.pyplot as plt

# GNN AUFGABE 5
# Ali Babaoglu 1827133

# This class represents a recurrent neural network with 2 neurons . 
class RNN(object):
    # initalizations of weights, inputs and activation function
    def __init__(self):
        self.inputs = np.array([0.0, 0.0, 1])
        self.weights =np.array([[-4, 1.5, -3.37], [-1.5, 0, 0.125]])
        self.activation_function= lambda x: 2 / (1 + np.exp(-2 * x)) - 1
    
    def activate(self):
        """
        performs a matrix multiplication and then applies the tangent hyperbolic transfer function

        Returns: output of neurons
        
        """
        self.out = np.array(np.dot(self.weights,self.inputs))
        self.out=  self.activation_function(self.out)
        self.inputs = np.append(self.out, 1)
        return self.out

rnn = RNN()
output1 = list()
output2 = list()
print("Results:")
# four runs
for i in range(4):
    out = rnn.activate()
    print(out)
    output1 = np.append(output1, out[0])
    output2 = np.append(output2, out[1])
    
plt.plot(output1)
plt.plot(output2)
plt.show()
        
```
